What is Pre-Tax National Income ? <br>
Pretax, post-replacement national income, which for simplicity we often refer to as “pretax
income,” is equal to the sum of all pretax income flows accruing to the individual owners of the
production factors (labor and capital) before the operation of the tax and transfer system, but
after the operation of the social insurance system.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

sns.set(color_codes=True)
```

```python
data = pd.read_csv(r'/content/WID_Data_28012023-054957.csv',sep =";",skiprows=[0],encoding="utf-8")
data.head()
```

```python
data.columns
```

```python
data = data.rename(columns={'sptinc_z_AF\nPre-tax national income \nTop 10% | share\nAfghanistan':'Afghanistan',
                            'sptinc_z_IN\nPre-tax national income \nTop 10% | share\nIndia':"India",
                            'sptinc_z_DE\nPre-tax national income \nTop 10% | share\nGermany':"Germany"})
```

```python
data.head()
```

Data for INDIA

```python
data_India = data[["Percentile","Year","India"]]
```

India Top 10%

```python
in_top10 = data_India[data_India["Percentile"]=="p90p100"]
in_top10
```

India Bottom 50%

```python
in_bot50 = data_India[data_India["Percentile"]=="p0p50"]
in_bot50
```

Plotting Graph between Top 10% and Bottom 50% India Pre-Tax Income

```python
sns.lineplot("Year","India",data=in_top10);
sns.lineplot("Year","India",data=in_bot50);
plt.tight_layout;
plt.legend(title="Groups",labels=["Top 10%(90-100)","Bottom Half (0-50)"]);
plt.ylabel("Pre-Tax National Income");
plt.xlabel("Year");
fig=plt.gcf();
fig.set_size_inches(12,10);
```
